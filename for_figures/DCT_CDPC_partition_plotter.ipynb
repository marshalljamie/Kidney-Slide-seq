```python
import pandas as pd
import numpy as np
import os
from os import path
from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn.metrics import pairwise_distances
import matplotlib.pyplot as plt
import seaborn as sns
import alphashape
from descartes import PolygonPatch
from shapely.geometry import Point
```

```python
celltypes = ['DCT','CDPC']
sections = ['medulla','cortex']
geno = 'human' # human, DKD, WT, UMOD-KI, UMOD-WT, where DKD refers to BTBR ob/ob and WT to BTBR wt/wt
# txt file listing unique ids for arrays of specified genotype
array_ids = list(pd.read_csv('to_process_all_{geno}.txt'.format(geno=geno.replace('-','')),header=None)[0])
for array_id in array_ids:
    for celltype in celltypes:
            # input_path is path to file with beads x features for all beads classified as celltype
            # features = {'barcode','x','y','clusters','section'}, where cluster is cluster assigned via kmeans clustering 
            # during partition assignment
            input_path = '{celltype}_info.csv'.format(celltype = celltype)
            
            if path.exists(input_path):
                cell_info = pd.read_csv(input_path,index_col=0)
                for section in sections:
                    cell_info_section = cell_info[cell_info['section']==section].copy()

                    out_path = '{array_id}_{celltype}_{section}_partitions.pdf'.format(celltype=celltype,section=section,array_id=array_id)
                    sns.set(rc={'figure.figsize':(5,5)})
                    sns.set_style("white")
                    p=sns.scatterplot(x=cell_info_section.x, y=cell_info_section.y, data=cell_info_section, hue=cell_info_section.clusters, palette='Spectral',alpha=0.3,ec='black',s=40,legend=False)
                    sns.despine(fig=None, ax=None, top=False, right=False, left=False, bottom=False, offset=None, trim=False)
                    p.set(xticks=[],yticks=[],xlabel=None,ylabel=None)
                    plt.xlim(0,6000)
                    plt.ylim(0,6000)
                    plt.savefig(out_path,dpi=300,bbox_inches='tight')
                    plt.show()
```
